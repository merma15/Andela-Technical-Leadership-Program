```python
m = pymorphy2.MorphAnalyzer()
m.parse('Снеговик')
```

```python
txt = ['озеро', 'озера', 'озером', 'озере']
t = m.parse(txt[3])[0]
print(t)
```

```python
print(t.normal_form)
```

```python
print(t.tag)
```

```python
word = "стол"
p = m.parse(word)[0]
print(p.inflect({'nomn'}).word)  # именительный падеж
```
